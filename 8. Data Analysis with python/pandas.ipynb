#### PANDAS - DATAFRAME AND SERIES


Pandas is a powerful data manipulation library in python, widely used for data analysis and data cleaning. It provides two primary data structures: series and dataframes.

A Series is a one -dimensional array-like object that can hold any datatype.It is similar to a column in a table, while a dataframe is a two-dimensional,size-mutable, and potentially heterogenous tabular data structure while labeled axes (rows and columns)

```python
## MERGING AND JOINING DFS

df1=pd.DataFrame({'key':['A','B','C'],'value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
